```python
import pandas as pd


ind_bilateral = pd.read_excel("IND_RCEP_experiment_bilateral_results.csv.xlsx")
only_bilateral = pd.read_excel("ONLY_RCEP_experiment_bilateral_results.csv.xlsx")


ind_bilateral.columns = ind_bilateral.columns.str.strip().str.lower().str.replace(" ", "_")
only_bilateral.columns = only_bilateral.columns.str.strip().str.lower().str.replace(" ", "_")

print("Columns found:", list(ind_bilateral.columns))
```

```python

exp_col = [c for c in ind_bilateral.columns if "export" in c][0]
imp_col = [c for c in ind_bilateral.columns if "import" in c][0]
trade_col = [c for c in ind_bilateral.columns if "trade" in c and ("ratio" in c or "change" in c)][0]

print("Exporter column:", exp_col)
print("Importer column:", imp_col)
print("Trade change column:", trade_col)
```

```python
india_in_pairs = ind_bilateral[(ind_bilateral[exp_col] == "IND") | (ind_bilateral[imp_col] == "IND")]
india_out_pairs = only_bilateral[(only_bilateral[exp_col] == "IND") | (only_bilateral[imp_col] == "IND")]

print("India Inside pairs:", india_in_pairs.shape[0])
print("India Outside pairs:", india_out_pairs.shape[0])
```

```python
# Compute medians
median_trade_in = india_in_pairs[trade_col].median()
median_trade_out = india_out_pairs[trade_col].median()
net_trade_impact = median_trade_in - median_trade_out

# Display result
print("\n=== Bilateral Trade Median Analysis ===")
print(f"Median Trade Change (India Outside): {median_trade_out:.4f}")
print(f"Median Trade Change (India Inside):  {median_trade_in:.4f}")
print(f"Net Trade Benefit of Joining RCEP:   {net_trade_impact:.4f}")

# Interpretation
if net_trade_impact > 0:
    print("\n India gains in trade connectivity (positive trade reallocation).")
else:
    print("\n India faces trade diversion (negative impact).")
```

```python
import pandas as pd

summary = pd.DataFrame({
    "Scenario": ["India Outside", "India Inside", "Net Impact"],
    "Median_Trade_Change": [median_trade_out, median_trade_in, net_trade_impact],
    "Interpretation": [
        "Trade diversion loss (India excluded)",
        "Trade reallocation gain (India included)",
        "Net benefit of joining RCEP" if net_trade_impact > 0 else "Net loss from joining RCEP"
    ]
})

summary.to_excel("India_Bilateral_Trade_Impact.xlsx", index=False)
print("\n📘 Saved result as 'India_Bilateral_Trade_Impact.xlsx'")
```
