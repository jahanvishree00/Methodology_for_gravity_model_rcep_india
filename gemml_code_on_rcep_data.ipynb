```python
merged_data2 = pd.read_excel("RCEP Data.xlsx", sheet_name="RCEP Data (1)")
print(merged_data2.dtypes)
```

```python
numeric_cols = merged_data2.select_dtypes(include=[np.number]).columns
merged_data2[numeric_cols] = merged_data2[numeric_cols].round(4)
merged_data = merged_data2.copy()
```

```python
gme_data = gme.EstimationData(merged_data,imp_var_name = 'importer', exp_var_name = 'exporter',
                              year_var_name = 'year', trade_var_name = 'trade_flow')
```

```python
gme_model2 = gme.EstimationModel(
    gme_data,
    lhs_var="adjusted_trade_flow",
    rhs_var=[  "log_P","Exporter_GDP","E ", "Distance", "Trade_Creation","Import_Trade_Diversion","Export_Trade_Diversion","Tariff", "common_language" , "contiguity", "com_col"],
    fixed_effects=[["exporter", "year"], ["importer", "year"],["importer","exporter"]] # Adding country-pair fixed effects
)
```

```python
gme_model2.estimate()
```

```python
import pandas as pd

# Access the full results summary for "all" key (adjust if the key is different)
full_results = gme_model2.results_dict["all"]

# Convert the summary table to a DataFrame, including coefficients, p-values, standard errors, etc.
# Assuming `summary()` returns a table-like object, try accessing these attributes:
params_df = pd.DataFrame({
    "Coefficient": full_results.params,
    "Std_Error": full_results.bse,
    "P_value": full_results.pvalues,
    "Confidence_Interval_Lower": full_results.conf_int().iloc[:, 0],
    "Confidence_Interval_Upper": full_results.conf_int().iloc[:, 1]
})

# Export to CSV
params_df.to_csv("gme_full_results.csv", index_label="Parameter")
```

```python
# Look for RTA-related parameters
rta_candidates = [col for col in full_results.params.index if 'rta' in col.lower() or 'rcep' in col.lower()]
print("RTA candidates:", rta_candidates)

# Most likely matches (pick the right one):
possible_names = ['rta', 'RTA', 'rta_rcep', 'RTA_RCEP', 'rcep', 'RCEP', 'Trade_Creation']
for name in possible_names:
    if name in full_results.params.index:
        print(f"Found RTA parameter: '{name}' = {full_results.params[name]:.4f}")
        rta_coef = full_results.params[name]
        pti_yes = np.exp(rta_coef)
        print(f"PTI(Yes) = {pti_yes:.3f}")
        break
else:
    print("No RTA parameter found - check your model specification!")
```
